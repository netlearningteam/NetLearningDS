```python
import json
import matplotlib.pyplot as plt 
from PIL import Image
```

```python
def read_json(json_dir):
    with open(json_dir,'r') as load_f:
        load_dict = json.load(load_f)
    return load_dict


def output_dict(in_dict):
    for key,values in in_dict.items() : 
        print(key + ": ")
        print(values)
```

```python
ori_data_loc = "DS1/ori_data/train_data.json"
preprocess_data_loc = "DS1/preprocessd_data/train_data.json"
data_petri_pic_loc = "visualization_for_ds1/data1(petri).png" 
data_arr_pic_loc = "visualization_for_ds1/data1(arri).png" 
```

```python
ori_data = read_json(ori_data_loc)
print("*"*30,"raw data","*"*30)
output_dict(ori_data['data1'])
print("*"*30,"preprocess data","*"*30)
preprocess_data  = read_json(preprocess_data_loc)
output_dict(preprocess_data['data1'])
```

```python
im = Image.open(data_petri_pic_loc)
plt.figure(figsize=(60,40))
plt.imshow(im)
plt.axis('off') 
plt.title('Petri Net') 
plt.show()
```

```python
im = Image.open(data_arr_pic_loc)
plt.figure(figsize=(60,40))
plt.imshow(im)
plt.axis('off') 
plt.title('Reachable Graph') 
plt.show()
```
